```python
import os
```

```python
home = os.path.expanduser('~')
# Assign your workspace path (Change to your desired directory)
workspace = os.path.join(home,"vitis_ws")
comp_name = "zynq_zed"
app_name = "zynq_zed_leds_app"
bin_dir = "./bin"
```

```python
# 1. Locate your compiled bare-metal application binary (ELF)
app_elf = f"{workspace}/{app_name}/build/{app_name}.elf"

# 2. Locate the First Stage Bootloader (FSBL) automatically built by your platform component
fsbl_elf = f"{workspace}/{comp_name}/export/zynq_zed/sw/boot/fsbl.elf"

# 3. Locate your hardware FPGA Bitstream extracted from your XSA file
fpga_bit = f"{workspace}/{comp_name}/export/zynq_zed/hw/sdt/zynq_zed_ex.bit"

# Define the configuration file output location
bif_path = f"{bin_dir}/{app_name}.bif"

# Create the BIF configuration text block for a Zynq-7000 architecture
bif_content = f"""
the_ROM_image: {{
    [bootloader] {fsbl_elf}
    {fpga_bit}
    {app_elf}
}}
"""
```

```python
# Write the file to disk
with open(bif_path, "w") as bif_file:
    bif_file.write(bif_content.strip())

print(f"Boot configuration file written successfully to: {bif_path}")
```

```python
# 1. Update the path for this execution block
os.environ["PATH"] = f"/opt/Xilinx/Vitis/2024.1/bin:{os.environ['PATH']}"

# 2. Construct the exact command string using Python f-strings
cmd = f"bootgen -image {bif_path} -arch zynq -o {bin_dir}/BOOT.bin -w on"

# 3. Execute it and print the output
os.system(cmd)
```
