```python
import os
import vitis
import shutil
```

```python
# 1. Initialize the headless Vitis server engine
client = vitis.create_client()
```

```python
# 2. Assign your workspace path (Change to your desired directory)
home = os.path.expanduser('~')
ws_path = os.path.join(home,"vitis_ws")
client.set_workspace(path=ws_path)
```

```python
# 1. Capture the exact component list from your log
existing_components = client.list_components()

# 2. Extract just the text string names from the dictionaries
existing_names = [comp['name'] for comp in existing_components]

platform_name = "zynq_zed"

# 3. Check and load or create dynamically
if platform_name in existing_names:
    print(f"Platform '{platform_name}' already exists. Loading component...")
    platform_comp = client.get_component(name=platform_name)
else:
    platform_comp = client.create_platform_component (
        name=platform_name,
        hw_design="../vivado/bin/zynq_zed_ex.xsa",
        os="standalone",
        cpu="ps7_cortexa9_0" # Native CPU core for ZedBoard Zynq-7000
    )
```

```python
# Compile and build the platform configuration
platform_comp.build()
```

```python
ps7_init_tcl_path = os.path.join(ws_path,platform_name,"export",platform_name,"hw","sdt","ps7_init.tcl")
fsbl_elf_path = os.path.join(ws_path,platform_name,"export",platform_name, "sw", "boot", "fsbl.elf")

ps7_dest_path = os.path.join("./tcl")
fsbl_dest_path = os.path.join("./bin")

os.makedirs(ps7_dest_path, exist_ok=True)
os.makedirs(fsbl_dest_path, exist_ok=True)

if os.path.exists(ps7_init_tcl_path):
    shutil.copy2(ps7_init_tcl_path, ps7_dest_path)
    print(f"Success! {ps7_init_tcl_path} copied to : {ps7_dest_path}")
else:
    print(f"Warning: ps7_init.tcl not found at {ps7_init_tcl_path}.")

if os.path.exists(fsbl_elf_path):
    shutil.copy2(fsbl_elf_path, fsbl_dest_path)
    print(f"Success! {fsbl_elf_path} copied to : {fsbl_dest_path}")
else:
    print(f"Warning: ps7_init.tcl not found at {fsbl_elf_path}.")
```

```python
# Terminate the server connection and release the workspace lock
vitis.dispose()
```
